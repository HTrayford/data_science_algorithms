#### Description of Data Sources
**The New York Times COVID 19 Data** provides county-by-county information regarding COVID 19 pertinent data (number of infections, estimated deaths, etc) for 2021 from Jan 1, to Nov 1.  It is available for download at the GitHub site listed above. Fields in this data set include the Date, State, County Name, Number of Cases, Average Number of Cases per 100k People, Deaths, Average Deaths, and Average Deaths per 100k People. The most useful fields will be Date, State, County Name, Average Number of Cases per 100k and Average Deaths per 100k. This data set consists of nearly 1,000,000 rows of data encompassing one entry per county per day from January 1, to November 1, 2021.  Data is in the form of integers for case numbers and floats for averages.

The **tonmcg County Level Election Data** for the 2020 U.S. Presidential election provides information for each county in the U.S. during the 2020 Presidential election. It is available for download at the GitHub site listed above. Fields for each county include: State, County Name, Votes for the GOP, Votes for the Democratic Party, Total Votes, Difference in Votes, Percent of Votes for GOP, Percent of Votes for the Democratic Party, and the difference in the percentages. The fields of most use during this analysis include the State, County Name, Votes for the GOP and Votes for the Democratic Party. All other fields can be derived from these. This data set consist of approximately 3500 entries, one for each county (or county equivalent municpal entity) in the U.S. Data is in the form of integers for vote totals and floats for percentages.

The **U.S. County Population** and **Land Area Data** can be used to gather information about the population of each county as well as the population density. They are available for download at the census.gov websites listed above. Use of these data sets will require the assumption that the 2019 population estimates are reasonably accurate and that the land area of the counties hasn't changed meaningfully since the 2010 census.  These data sets consist of approximately 3500 entries, one for each county (or county equivalent municpal entity) in the U.S. Population is in the form of integers for number of persons, while Land Area is in the form of floats for square miles of land area.

By combining the first two data sets, it is possible to analyze how strong a correlation exists between a county's voting and how badly it has suffered during the COVID 19 pandemic. Similar analyses have been done showing that such a correlation exists, but popular media have not discussed the strength of this correlation nor attempted to make any predictions based on this information (e.g. what is the probability that a person contracts COVID 19 based on how strongly that person's county voted for each candidate), nor has there been discussion regarding the importance of voting information relative to population or population density, more traditionally important factors in the spread of disease.

#### Goals
The primary goal is to determine if it is possible to assess (and ultimately predict) risk of infection based on county voting data. A secondary goal is to assess if voting is a stronger predictor of risk than population or population density. Understanding the relative impact could inform policy makers within local governments as well as healthcare entities.

#### Type of Analysis

This will be a **regression** analysis to determine the impact of voting (how strongly in favor of one candidate over another) has on risk of infection.

#### Possible Sources of Difficulty

1) There are a number of caveats associated with the COVID-19 data. In particular, the way the COVID-19 data was reported changed in various, relatively minor ways, over the course of 2021. More importantly there are special cases for some regions. One important detail is that not all regions are consistently reported at the county level and this may increase the difficulty of performing the county level analysis for these regions. A second important detail is that even a cursory review of the data makes it clear that not all regions reported their data accurately (i.e. large number of infected persons and zero deaths). A third potential challenge is that this dataset includes nearly 1,000,000 rows of data and it may prove to be too large to use without paring it down. Details of these caveats and special cases are posted at the GitHub website for the data.

2) The county level election data also includes some caveats, most prominently that not all states have counties and not all states report their election data at a county-equivalent level. This will introduce some difficulty for these areas.

3) Difficulties associated with the census data are two-fold. Firstly, the 2019 population estimates are estimates based on the 2010 census as this data was released prior to the 2020 census. Secondly, the county land area data is also based on the 2010 census, but no updated estimates are provided. So we will be forced to apply 2019 population estimates to 2021 COVID-19 data and spread that over 2010 county land areas.

### Data Cleanup (EDA)

Data clean up was performed directly in the CSV files or the Excel/Numbers files prior to uploading them to the notebook. Data clean up consisted of:
- Removing unwanted columns (e.g. empty columns, population estimates by county for years other than 2019, Land Area data for each county other than the primary data point, etc)
- Modifying County name columns to include only the county name (i.e. removing the state name or abbreviation from the county name).
- Adding a "State" column so the data layout matches more closely with the COVID data.
- For the Population and Land Area data sets, rows for the entire United States and for whole States were removed, leaving just county (or county equivalent) data.

Additional cleanup will consist of removing NaN values (exclusively exist in the COVID data), evaluating the validity of specific COVID entries (e.g. rows with large numbers of cases but zero deaths) and verifying the county names match across data sources to enable merging the data.

#### Clean County names to establish commonality between the four data sets

```python
#Clean the county name columns from each of the three smaller data sets to make them match the county names
#as listed in the COVID-19 dataset

def cleanCountyName (dataframe, county_header):
    for index in dataframe.index:
        county_name = dataframe.loc[index, county_header]
        if 'County' in county_name:
            dataframe.loc[index, county_header] = county_name.replace('County', '', 1).strip()
        elif 'Parish' in county_name:
            dataframe.loc[index, county_header] = county_name.replace('Parish', '', 1).strip()
        elif 'city' in county_name:
            dataframe.loc[index, county_header] = county_name.replace('city', '', 1).strip()
        elif county_name == 'New York City':
            dataframe.loc[index, county_header] = 'New York'

        if 'La Salle' in county_name:
            dataframe.loc[index, county_header] = 'LaSalle'
        elif 'DoÃ±a Ana' in county_name:
            dataframe.loc[index, county_header] = 'Dona Ana'

cleanCountyName(election_df, 'county_name')
cleanCountyName(population_df, 'County')
cleanCountyName(landarea_df, 'County')
cleanCountyName(covid_df, 'county')
```

#### Clean State names to remove extraneous white space and to correct spelling

```python
#Clean State name columns from each of the three smaller data sets to make them match the county names
#as listed in the COVID-19 dataset

#For some reason, this cell takes ~5 minutes to run (way, way longer than the previous cell, which should be about the same)

def cleanStateName (dataframe, state_header):
    for index in dataframe.index:
        state_name = dataframe.loc[index, state_header]
        dataframe.loc[index, state_header] = dataframe.loc[index, state_header].strip()
        if dataframe.loc[index, state_header] == 'Deleware':
            dataframe.loc[index, state_header] = 'Delaware'
    
cleanStateName(election_df, 'state_name')
cleanStateName(population_df, 'State')
cleanStateName(landarea_df, 'State')
cleanStateName(covid_df, 'state')
```

#### Remove Alaska from the data set because there is little or no correlation between COVID reporting regions and voting regions

```python
#Remove Alaska because its voting districts have no relationship to its County equivalent structure
election_df = election_df[election_df.state_name != 'Alaska']
population_df = population_df[population_df.State != 'Alaska']
landarea_df = landarea_df[landarea_df.State != 'Alaska']
covid_df = covid_df[covid_df.state != 'Alaska']
```

#### The COVID data includes parts of the world which cannot vote, so they need to be removed

```python
#Remove places from COVID data which do not appear in any of the other datasets
covid_df = covid_df[covid_df.state != 'Puerto Rico']
covid_df = covid_df[covid_df.state != 'Northern Mariana Islands']
covid_df = covid_df[covid_df.state != 'Virgin Islands']

#Remove COVID entries whose "county" value is "unknown" since these cannot be mapped to any of the other data
covid_df = covid_df[covid_df.county != 'Unknown']
```

#### Add Oglala Lake County to the Land Area data since it is only missing from this one data set

```python
#Add Oglala Lakota land area to the landarea_df because it is missing. Value of 2094 obtained from:
#https://en.wikipedia.org/wiki/Oglala_Lakota_County,_South_Dakota
new_row = pd.DataFrame.from_dict({'County': ['Oglala Lakota'], 'State': ['South Dakota'], 'Area (sq. mi.)': ['2094.0']})
landarea_df = landarea_df.append(new_row, ignore_index = True);
```

#### Kalawao is present in all data sets except voting, this cannot be recovered, so remove Kalawao from all data sets

```python
#Remove Kalawao from land area, population, and COVID data because it does not appear in the election data set
population_df = population_df[population_df.County != 'Kalawao']
landarea_df = landarea_df[landarea_df.County != 'Kalawao']
covid_df = covid_df[covid_df.county != 'Kalawao']

#Remove locations which appear in landarea but nowhere else
landarea_df = landarea_df[landarea_df.County != 'Yellowstone National Park']
landarea_df = landarea_df[((landarea_df.State != 'South Dakota') | (landarea_df.County != 'Shannon'))]
landarea_df = landarea_df[((landarea_df.State != 'Virginia') | (landarea_df.County != 'Clifton Forge'))]
landarea_df = landarea_df[((landarea_df.State != 'Virginia') | (landarea_df.County != 'South Boston'))]
```

#### The Population data set includes commas for numbers > 999, removing these allows Population to be treated as an integer vice a string

```python
#Clean Population numbers to remove commas
for index in population_df.index:
    population = population_df.loc[index, 'Population']
    population_df.loc[index, 'Population'] = population.replace(',','')
```

#### Group each data set by State, then by County (or equivalent) in order to consolidate numbers and ensure we can properly compare county-by-county data

```python
#Group all four data sets to align them with the same set of keys in preparation
#for combining the desired columns into a single dataframe

election_df = election_df.astype({'votes_gop':'int', 'votes_dem':'int', 'total_votes':'int', 'diff':'int', 'per_gop':'float', 'per_dem':'float', 'per_point_diff':'float'})
grouped_gop = election_df.groupby(['state_name', 'county_name'])['votes_gop'].sum() #sums results of just 1 row => no change
grouped_dem = election_df.groupby(['state_name', 'county_name'])['votes_dem'].sum() #sums results of just 1 row => no change

population_df = population_df.astype({'Population':'int'})
grouped_population = population_df.groupby(['State', 'County'])['Population'].sum()

landarea_df = landarea_df.astype({'Area (sq. mi.)':'float'})
grouped_landarea = landarea_df.groupby(['State', 'County'])['Area (sq. mi.)'].sum()

covid_df = covid_df.astype({'cases': 'int', 'cases_avg': 'float', 'cases_avg_per_100k':'float', 'deaths':'int', 'deaths_avg':'float', 'deaths_avg_per_100k':'float'})
grouped_cases = covid_df.groupby(['state', 'county'])['cases'].sum()
grouped_deaths = covid_df.groupby(['state', 'county'])['deaths'].sum()

# Want to create a dataframe that has the following columns:
# state, county, case_count, death_count, population, population_density
```

#### The NY Times specifically indicates Joplin, MO and Kansas City, MO are reported separately from their respective Counties, we need to reverse this

```python
#Combine COVID cases for the following cities with the data for the surrounding County
grouped_cases.loc[('Missouri','Jasper')] = grouped_cases.loc[('Missouri','Jasper')] + grouped_cases.loc[('Missouri','Joplin')]
grouped_cases.loc[('Missouri','Jackson')] = grouped_cases.loc[('Missouri','Jackson')] + grouped_cases.loc[('Missouri','Kansas City')]

#Now remove the COVID cases associated with these cities
grouped_cases.drop(('Missouri', 'Joplin'), axis=0, inplace=True)
grouped_cases.drop(('Missouri', 'Kansas City'), axis=0, inplace=True)
```

```python
#same as above, but for deaths
grouped_deaths.loc[('Missouri','Jasper')] = grouped_deaths.loc[('Missouri','Jasper')] + grouped_deaths.loc[('Missouri','Joplin')]
grouped_deaths.loc[('Missouri','Jackson')] = grouped_deaths.loc[('Missouri','Jackson')] + grouped_deaths.loc[('Missouri','Kansas City')]

#Now remove the COVID deaths associated with these cities
grouped_deaths.drop(('Missouri', 'Joplin'), axis=0, inplace=True)
grouped_deaths.drop(('Missouri', 'Kansas City'), axis=0, inplace=True)
```

#### The NY Times specifically indicates that the data for the 5 burroughs of New York are consolidated into a single entry, we need to reverse this. As an approximation, we are distributing the cases and deaths for New York City evenly across the 5 burroughs.

```python
#Divide COVID cases for New York City by 5 and distribute evenly among the 5 borroughs of the city to match reporting for other data sets
total_NY_cases = grouped_cases.loc[('New York','New York')]
cases_per_borrough = total_NY_cases // 5
grouped_cases[('New York', 'New York')] = cases_per_borrough
grouped_cases[('New York', 'Bronx')] = cases_per_borrough
grouped_cases[('New York', 'Kings')] = cases_per_borrough
grouped_cases[('New York', 'Queens')] = cases_per_borrough
grouped_cases[('New York', 'Richmond')] = cases_per_borrough
```

```python
#Same as above, but for deaths
total_NY_deaths = grouped_deaths.loc[('New York','New York')]
deaths_per_borrough = total_NY_deaths // 5
grouped_deaths[('New York', 'New York')] = deaths_per_borrough
grouped_deaths[('New York', 'Bronx')] = deaths_per_borrough
grouped_deaths[('New York', 'Kings')] = deaths_per_borrough
grouped_deaths[('New York', 'Queens')] = deaths_per_borrough
grouped_deaths[('New York', 'Richmond')] = deaths_per_borrough
```

#### Build the data frame we'll use to create visualizations and begin the analysis

```python
consolidated_dictionary = {'State':[], 'County':[], 'GOP':[], 'DEM':[], 'Percent GOP':[], 'Percent DEM':[], 'Population':[], 'Area':[], 'Population Density':[], 'Cases':[], 'Deaths':[]}
for keys in cases_keys:
    consolidated_dictionary['State'].append(keys[0])
    consolidated_dictionary['County'].append(keys[1])
    consolidated_dictionary['GOP'].append(grouped_gop[keys])
    consolidated_dictionary['DEM'].append(grouped_dem[keys])
    consolidated_dictionary['Percent GOP'].append(grouped_gop[keys] / (grouped_gop[keys] + grouped_dem[keys]))
    consolidated_dictionary['Percent DEM'].append(grouped_dem[keys] / (grouped_gop[keys] + grouped_dem[keys]))
    consolidated_dictionary['Population'].append(grouped_population[keys])
    consolidated_dictionary['Area'].append(grouped_landarea[keys])
    consolidated_dictionary['Population Density'].append(grouped_population[keys] / grouped_landarea[keys]) #This crashed with divide by zero?!?
    consolidated_dictionary['Cases'].append(grouped_cases[keys])
    consolidated_dictionary['Deaths'].append(grouped_deaths[keys])

consolidated_df = pd.DataFrame.from_dict(consolidated_dictionary)
consolidated_df.head()
```

## Model Selection and Analysis

### Forward Stepwise Selection

We will consider GOP, Percent GOP, Population, Population Density, and Area as potential predictors. DEM and Percent DEM are obviously colinear (anti-linear actually), so will be excluded.

#### Forward Stepwise Selection of a model for Cases

##### Zero Predictor Model - Cases

```python
mean_cases_array = np.full((1, 3106), normed_df['Cases'].mean())
cases0 = smf.ols(formula="Cases~Mean", data=pd.DataFrame({'Cases':normed_df['Cases'], 'Mean':mean_cases_array[0]})).fit()
cases0.rsquared_adj
```

The Adjusted $R^2$ for the zero predictor model is effectively zero. Next we'll add one predictor at a time to find the one that results in the largest.

##### One Predictor Models - Cases

```python
cases1_gop = smf.ols(formula='Cases~GOP', data=normed_df).fit()
print('Adj R^2 for GOP:', cases1_gop.rsquared_adj)

cases1_percent_gop = smf.ols(formula='Cases~Percent_GOP', data=normed_df).fit()
print('Adj R^2 for Percent GOP:', cases1_percent_gop.rsquared_adj)

cases1_pop = smf.ols(formula='Cases~Population', data=normed_df).fit()
print('Adj R^2 for Population:', cases1_pop.rsquared_adj)

cases1_pop_den = smf.ols(formula='Cases~Population_Density', data=normed_df).fit()
print('Adj R^2 for Population Density:', cases1_pop_den.rsquared_adj)

cases1_area = smf.ols(formula='Cases~Area', data=normed_df).fit()
print('Adj R^2 for Area:', cases1_area.rsquared_adj)
```

The Adjusted $R^2$ for Percent GOP is the highest, so we will select that predictor and look at how the model can be improved by adding a second predictor

##### Two Predictor Models - Cases

```python
cases2_gop = smf.ols(formula='Cases~Percent_GOP+GOP+Percent_GOP*GOP', data=normed_df).fit()
#print(cases2_gop.summary())
print('Adj R^2 for GOP:', cases2_gop.rsquared_adj, '; GOP p-value:', cases2_gop.pvalues[2], '; GOP*Percent_GOP p-value:', cases2_gop.pvalues[3])

cases2_pop = smf.ols(formula='Cases~Percent_GOP+Population+Percent_GOP*Population', data=normed_df).fit()
print('Adj R^2 for Population:', cases2_pop.rsquared_adj, '; Population p-value:', cases2_pop.pvalues[2], '; Population*Percent_GOP p-value:', cases2_pop.pvalues[3])

cases2_pop_den = smf.ols(formula='Cases~Percent_GOP+Population_Density+Percent_GOP*Population_Density', data=normed_df).fit()
print('Adj R^2 for Pop Density:', cases2_pop_den.rsquared_adj, '; Pop Den p-value:', cases2_pop_den.pvalues[2], '; Pop Den*Percent_GOP p-value:', cases2_pop_den.pvalues[3])

cases2_area = smf.ols(formula='Cases~Percent_GOP+Area+Percent_GOP*Area', data=normed_df).fit()
print('Adj R^2 for Area:', cases2_area.rsquared_adj, '; Area p-value:', cases2_area.pvalues[2], '; Area*Percent_GOP p-value:', cases2_area.pvalues[3])
```

The largest $R^2$ for this round of selection is for GOP. The 95% confidence interval for GOP does not include zero and its p-value is effectively zero, so we conclude that it is a beneficial predictor and will select it.

##### Three Predictor Models - Cases

```python
#Percent_GOP, GOP

cases3_pop = smf.ols(formula='Cases~Percent_GOP+GOP+Population+Percent_GOP*GOP+Percent_GOP*Population+GOP*Population', data=normed_df).fit()
#print(cases3_pop.summary())
print('Adj R^2 for Population:', cases3_pop.rsquared_adj)

cases3_pop_den = smf.ols(formula='Cases~Percent_GOP+GOP+Population_Density+Percent_GOP*GOP+Percent_GOP*Population_Density+GOP*Population_Density', data=normed_df).fit()
#print(cases3_pop_den.summary())
print('Adj R^2 for Population Density:', cases3_pop_den.rsquared_adj)

cases3_area = smf.ols(formula='Cases~Percent_GOP+GOP+Area+Percent_GOP*GOP+Percent_GOP*Area+GOP*Area', data=normed_df).fit()
#print(cases3_area.summary())
print('Adj R^2 for Area:', cases3_area.rsquared_adj)
```

The model with the highest Adjusted $R^2$ is the model which includes the raw population. In this model, the Percent GOP:Population and the GOP:Population terms are not statistically significant and will be excluded despite choosing the Population model.

##### Four Predictor Models - Cases

```python
#Percent_GOP, GOP, Population

cases4_pop_den = smf.ols(formula='Cases~Percent_GOP+GOP+Population+Population_Density+Percent_GOP*GOP+Percent_GOP*Population_Density+GOP*Population_Density+Population*Population_Density', data=normed_df).fit()
#print(cases4_pop_den.summary())
print('Adj R^2 for Population Density:', cases4_pop_den.rsquared_adj)

cases4_area = smf.ols(formula='Cases~Percent_GOP+GOP+Population+Area+Percent_GOP*GOP+Percent_GOP*Population+Percent_GOP*Area+GOP*Population+GOP*Area+Population*Area', data=normed_df).fit()
#print(cases4_area.summary())
print('Adj R^2 for Area:', cases4_area.rsquared_adj)
```

The adjusted $R^2$ with the Population Density term is larger than without it and is larger than the $R^2$ obtained by adding the Area predictor.  Further, looking at the summary information, we find the p-value for all coefficients to be zero. However, the GOP:Population_Density and Population:Population_Density terms are effectively zero (< 9e-07) and will be left out of further analysis.

##### Five Predictor Model - Cases

```python
#Percent_GOP, GOP, Population, Population Density

cases5_area = smf.ols(formula='Cases~Percent_GOP+GOP+Population+Population_Density+Area+Percent_GOP*GOP+Percent_GOP*Population+Percent_GOP*Population_Density+Percent_GOP*Area+GOP*Population+GOP*Area+Population*Area+Population_Density*Area', data=normed_df).fit()
#print(cases5_area.summary())
print('Adj R^2 for Area:', cases5_area.rsquared_adj)
```

This final model, which includes all five predictors has the same Adjusted $R^2$ as the previous model with excluded the Area predictor, so this model will definitely not be selected since there is no improvement with the added complexity.

##### Plot of Adjusted $R^2$ vs. Number of Predictors - Cases

```python
cases_predictor_count = [0, 1, 2, 3, 4, 5]
cases_r2_adj = [cases0.rsquared_adj, cases1_percent_gop.rsquared_adj, cases2_gop.rsquared_adj, cases3_pop.rsquared_adj, cases4_pop_den.rsquared_adj, cases5_area.rsquared_adj]

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(9, 4))
axes.plot(cases_predictor_count, cases_r2_adj);
axes.set_title('Adjusted R^2 vs. Number of Predictors');
axes.set_xlabel('Number of Predictors');
axes.set_ylabel('R^2');
```

#### Analysis: Forward Stepwise Selection - Cases

The plot of Adjusted $R^2$ versus Number of Predictors shows the diminishing returns for increasing complexity of the model. Including the first two predictors results in a fairly substantial increase in Adjusted $R^2$ from nearly zero to about 0.08. Including the next two predictors, and thereby doubling the model's complexity, results in a further increase in Adjusted $R^2$ of only about 0.02.

The first included predictor is Percent GOP and the second predictor is Raw GOP votes, meaning the more strongly a county voted for the GOP and the more GOP voters there are in a County, the more Cases that County has had during the study period.  Also accounting for total number of people in a County and the density of that population increases the accuracy, but the interpretive complexity rises rapidly.

#### Repeat Forward Stepwise Selection to find a model for Deaths

##### Zero Predictor Model - Deaths

```python
mean_deaths_array = np.full((1, 3106), normed_df['Deaths'].mean())
deaths0 = smf.ols(formula="Deaths~Mean", data=pd.DataFrame({'Deaths':normed_df['Deaths'], 'Mean':mean_deaths_array[0]})).fit()
deaths0.rsquared_adj
```

The Adjusted $R^2$ for the zero predictor model is effectively zero. Next we'll add one predictor at a time to find the one that results in the largest.

##### One Predictor Models - Deaths

```python
deaths1_gop = smf.ols(formula='Deaths~GOP', data=normed_df).fit()
print('Adj R^2 for GOP:', deaths1_gop.rsquared_adj)

deaths1_percent_gop = smf.ols(formula='Deaths~Percent_GOP', data=normed_df).fit()
print('Adj R^2 for Percent GOP:', deaths1_percent_gop.rsquared_adj)

deaths1_pop = smf.ols(formula='Deaths~Population', data=normed_df).fit()
print('Adj R^2 for Population:', deaths1_pop.rsquared_adj)

deaths1_pop_den = smf.ols(formula='Deaths~Population_Density', data=normed_df).fit()
print('Adj R^2 for Population Density:', deaths1_pop_den.rsquared_adj)

deaths1_area = smf.ols(formula='Deaths~Area', data=normed_df).fit()
print('Adj R^2 for Area:', deaths1_area.rsquared_adj)
```

As with Cases, the strongest single predictor model is the one which uses Percent GOP, so we will select that and move to two predictor models.

##### Two Predictor Models - Deaths

```python
deaths2_gop = smf.ols(formula='Deaths~Percent_GOP+GOP+Percent_GOP*GOP', data=normed_df).fit()
#print(deaths2_gop.summary())
print('Adj R^2 for GOP:', deaths2_gop.rsquared_adj, '; GOP p-value:', deaths2_gop.pvalues[2], '; GOP*Percent_GOP p-value:', deaths2_gop.pvalues[3])

deaths2_pop = smf.ols(formula='Deaths~Percent_GOP+Population+Percent_GOP*Population', data=normed_df).fit()
print('Adj R^2 for Population:', deaths2_pop.rsquared_adj, '; Population p-value:', deaths2_pop.pvalues[2], '; Population*Percent_GOP p-value:', deaths2_pop.pvalues[3])

deaths2_pop_den = smf.ols(formula='Deaths~Percent_GOP+Population_Density+Percent_GOP*Population_Density', data=normed_df).fit()
print('Adj R^2 for Pop Density:', deaths2_pop_den.rsquared_adj, '; Pop Den p-value:', deaths2_pop_den.pvalues[2], '; Pop Den*Percent_GOP p-value:', deaths2_pop_den.pvalues[3])

deaths2_area = smf.ols(formula='Deaths~Percent_GOP+Area+Percent_GOP*Area', data=normed_df).fit()
print('Adj R^2 for Area:', deaths2_area.rsquared_adj, '; Area p-value:', deaths2_area.pvalues[2], '; Area*Percent_GOP p-value:', deaths2_area.pvalues[3])
```

Unlike any of the models for Cases, the best two predictor model for Deaths is one that uses Percent GOP and Area. Interestingly, the p-values for the other three predictors in this model are all > 0.05 and likely not statistically significant.  Additionally, even the best two predictor model only saw an increase in Adjusted $R^2$ of about 0.001, substantially smaller than the increase seen for the two predictor model for Cases.  It is likely that other than Percent GOP there are no other predictors (of those under consideration here) for Deaths in a County which are important to include in the model.

##### Three Predictor Models - Deaths

```python
#Percent_GOP, Area

deaths3_pop = smf.ols(formula='Deaths~Percent_GOP+Area+Population+Percent_GOP*Area+Percent_GOP*Population+Area*Population', data=normed_df).fit()
#print(deaths3_pop.summary())
print('Adj R^2 for Population:', deaths3_pop.rsquared_adj)

deaths3_pop_den = smf.ols(formula='Deaths~Percent_GOP+Area+Population_Density+Percent_GOP*Area+Percent_GOP*Population_Density+Area*Population_Density', data=normed_df).fit()
#print(deaths3_pop_den.summary())
print('Adj R^2 for Population Density:', deaths3_pop_den.rsquared_adj)

deaths3_gop = smf.ols(formula='Deaths~Percent_GOP+Area+GOP+Percent_GOP*Area+Percent_GOP*GOP+Area*GOP', data=normed_df).fit()
#print(deaths3_gop.summary())
print('Adj R^2 for GOP:', deaths3_gop.rsquared_adj)
```

For the three predictor model for Deaths, the GOP predictor has the largest impact on Adjusted $R^2$, looking at the summary for this model shows that all the parameters which involve GOP are very nearly zero. Despite this, the model's performance is enhanced by including this parameter. Additionally, with this model the Area predictor now has a p-value of 0.241 and its 95% confidence interval spans zero, meaning it is no longer statistically significant. If we were using Bidirectional elimination, we would remove this predictor and proceed without it. Since we are using Forward Stepwise Selection, we will keep it.

##### Four Predictor Models - Deaths

```python
#Percent_GOP, Area, GOP

deaths4_pop = smf.ols(formula='Deaths~Percent_GOP+Area+GOP+Population+Percent_GOP*Area+Percent_GOP*GOP+Percent_GOP*Population+Area*GOP+Area*Population+GOP*Population', data=normed_df).fit()
#print(deaths4_pop.summary())
print('Adj R^2 for Population:', deaths4_pop.rsquared_adj)

deaths4_pop_den = smf.ols(formula='Deaths~Percent_GOP+Area+GOP+Population_Density+Percent_GOP*Area+Percent_GOP*GOP+Percent_GOP*Population_Density+Area*GOP+Area*Population_Density+GOP*Population_Density', data=normed_df).fit()
print(deaths4_pop_den.summary())
print('Adj R^2 for Population Density:', deaths4_pop_den.rsquared_adj)
```

These two models have essentially the same Adjusted $R^2$ value, the model which includes the Population predictor has 5 parameters whose p-values are > 0.05 while the model which includes the Population Density predictor only has two such paramters. Additionally, the model including Population Density has the higher Adjusted $R^2$ value. It will be the next predictor added to the model.

##### Five Predictor Model - Deaths

```python
deaths5_pop = smf.ols(formula='Deaths~Percent_GOP+Area+GOP+Population_Density+Population+Percent_GOP*Area+Percent_GOP*GOP+Percent_GOP*Population+Percent_GOP*Population_Density+Area*GOP+Area*Population+Area*Population_Density+GOP*Population+GOP*Population_Density+Population*Population*Population_Density', data=normed_df).fit()
#print(deaths5_pop.summary())
print('Adj R^2 for Population:', deaths5_pop.rsquared_adj)
```

This final model has a slightly smaller Adjusted $R^2$ value than the four predictor model which includes the Population Density.

### Cross Validation

#### Split Data into Training and Test subsets

```python
training_df = normed_df.iloc[::2, :] # Choose rows from 0 to the last row with a step of 2 (evens), choose columns from 0 to the last column
test_df = normed_df.iloc[1::2, :] # Choose rows from 1 to the last row with a step of 2 (odds), choose columns from 0 to the last column
```

#### Use Simple Linear Regression model for Cases and fit it to the training data

```python
cases_train = smf.ols(formula='Cases~Percent_GOP', data=training_df).fit()
print("R^2 for Cases Model using Training Data:", cases_train.rsquared)
```

The model has a slightly lower $R^2$ value using just the training data than the Adjusted $R^2$ value with the full data set. This wouldn't necessarily be the case, but it is here. We do expect the MSE this model has on the training data set to be lower than the MSE on the test data set.

#### Find the Mean Squared Error of this model relative to the data it was trained on

```python
cases_train_pred = cases_train.predict(training_df) / 100000

print("MSE of Cases Model using training data:", mean_squared_error(training_df['Cases'] / 100000, cases_train_pred))
```

#### Find MSE for this model on the Test data

```python
cases_test_pred = cases_train.predict(test_df) / 100000
print("MSE of Cases Model using Test data:", mean_squared_error(test_df['Cases'] / 100000, cases_test_pred))
```

As expected the MSE for the model is larger on the Test data than on the Training data.

#### Use Simple Linear Regression model for Deaths and fit it to the training Data

```python
deaths_train = smf.ols(formula='Deaths~Percent_GOP', data=training_df).fit()
print("R^2 for Deaths Model using Training Data:", deaths_train.rsquared)
```

#### Find the Mean Squared Error of this model relative to the data it was trained on

```python
deaths_train_pred = deaths_train.predict(training_df) / 100000

print("MSE of Deaths Model using training data:", mean_squared_error(training_df['Deaths'] / 100000, deaths_train_pred))
```

#### Find MSE for this model on the Test data

```python
deaths_test_pred = deaths_train.predict(test_df) / 100000
print("MSE of Deaths Model using Test data:", mean_squared_error(test_df['Deaths'] / 100000, deaths_test_pred))
```

As expected, the MSE for the Deaths model is lower than the MSE for the Cases model on both the training and the test data sets. Additionally, the Deaths model has a lower MSE on the training data than the test data.

#### Analysis of Cross Validaton

The fact that our models achieve a slightly higher $R^2$ when trained using the entire data set than they do when using half the data, combined with the relatively small changes in the MSE between training and test data, indicates that the Simple Linear Regression model is not significantly over fitting the data despite using the entire dataset for training.  This is expected given the relative inflexibility of Simple Linear Regression and the relatively large sample size and small number of predictors.

After reviewing models for Cases and Deaths using Simple Linear Regression, Forward Stepwise Selection, Polynomial (Quadratic and Cubic) models and doing cross validation on the Simple Linear Regression models it is clear that the 'Percent GOP' predictor has a meaningful impact on both infection and death rates in each County.
